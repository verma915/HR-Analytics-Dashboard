```python
# Check the data discrepancy
print("üîç INVESTIGATING LEFT EMPLOYEES WITHOUT LEAVING DATES:")

# Count employees with "Left" remark but no leaving date
left_employees = df[df['Remark'] == 'Left']
left_without_date = left_employees[left_employees['Leaving Date'].isna()]

print(f"Total employees with 'Left' remark: {len(left_employees)}")
print(f"Left employees WITHOUT leaving date: {len(left_without_date)}")
print(f"Left employees WITH leaving date: {len(left_employees) - len(left_without_date)}")

if len(left_without_date) > 0:
    print(f"\nüìù Samples of 'Left' employees without leaving dates:")
    for i, (idx, row) in enumerate(left_without_date.head(10).iterrows(), 1):
        print(f"{i}. {row['Name']} | Joined: {row['Joining Date']} | Remark: {row['Remark']}")

# Also check if there are leaving dates without "Left" remark
date_without_left = df[df['Leaving Date'].notna() & (df['Remark'] != 'Left')]
print(f"\nEmployees with Leaving Date but Remark ‚â† 'Left': {len(date_without_left)}")
```

```python
# Use ONLY reliable data - employees with actual leaving dates
df['Reliable_Status'] = 'Active'
df.loc[df['Leaving Date'].notna(), 'Reliable_Status'] = 'Left'

print("üìä RELIABLE STATUS (Only using employees with leaving dates):")
reliable_counts = df['Reliable_Status'].value_counts()
print(reliable_counts)
print(f"Reliable Attrition Rate: {(reliable_counts['Left'] / len(df) * 100):.1f}%")
```
